```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('house_data.csv')
```

```python
data
```

```python
data.rename(columns={'postcode':'outward_code'}, inplace=True)
```

```python
#reorder columns
cols = data.columns.to_list()

data = data[['price', 'living_rooms', 'bedrooms', 'bathrooms', 'town_city', 'outward_code', 'property_type', 'desc']]

data
```

```python
data.dropna(inplace=True)
```

```python
#changing price to float
data = data[data['price'] != 'POA']

data['price'] = [x.replace(',', '') for x in data['price']]

data['price'] = data['price'].astype(float)
```

```python
data
```

```python
#remove ' -' from property_type column
data['property_type'] = data['property_type'].str.replace(' -', '')
```

```python
#cleaning outward_code column
data['outward_code'] = data['outward_code'].replace({'London':np.nan, 'London,':np.nan, 'Dogs,':np.nan})
data.dropna()
data['outward_code'].unique()
```

```python
data['desc'].str.lower()
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data['desc'][5]
```

```python
data['desc'] = data.desc.str.lower()
```

```python
#remove special characters from desc
data['desc'] = data['desc'].str.replace(r'[^\w\s]+', '').str.strip()
```

```python
#remove numbers from desc
data['desc'] = data['desc'].str.replace('\d+', '')
```

```python
#remove special characters from outward_code
data['outward_code'] = data['outward_code'].str.replace(r'[^\w\s]+', '').str.strip()
```

```python
#finally, remove outliers; remove prices which are top-5-percentile 
percentile = data['price'].quantile(0.95)
```

```python
data = data[data['price'] <= percentile]
data
```

```python
#Output cleaned dataframe to csv
data.to_csv('cleaned_data')
```
